```python
import matplotlib.pyplot as plt
import cv2
import os

def show_images_vertical(image_list,
                         folder_path='/content/drive/MyDrive/pbas_27_04_2025/PBAS/results/eval/mvtec_wood'):

    images = []
    titles = []

    for image_name in image_list:
        image_path = os.path.join(folder_path, image_name)
        if not os.path.exists(image_path):
            print(f"{image_name} bulunamadı, atlanıyor.")
            continue
        img = cv2.imread(image_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        images.append(img)
        titles.append(image_name)

    n = len(images)
    if n == 0:
        print("Hiçbir görüntü bulunamadı.")
        return

    fig, axs = plt.subplots(n, 1, figsize=(8, 5 * n))

    if n == 1:
        axs.imshow(images[0])
        axs.set_title(titles[0])
        axs.axis('off')
    else:
        for i in range(n):
            axs[i].imshow(images[i])
            axs[i].set_title(titles[i])
            axs[i].axis('off')

    plt.tight_layout()
    plt.show()


# Örnek kullanım
show_images_vertical([
    "001.png",
    "002.png",
    "003.png",
    "004.png"
])
```
